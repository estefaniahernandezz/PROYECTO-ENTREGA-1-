```python
#Importamos librerías necesarias
import os
import pandas as pd #se importa la libreria pandas
import matplotlib.pyplot as plt

#Creamos la ruta base y la ruta al archivo trains.csv
basepath="/content/drive/MyDrive/ARCHIVOS IA"
trainpath=os.path.join(basepath,"train.csv")
datatrain=pd.read_csv(trainpath)

#Estudiantes por carrera

datatrain["E_PRGM_ACADEMICO"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 programas académicos con más estudiantes")
plt.xlabel("Carreras")
plt.ylabel("Estudiantes")
plt.show()

#Estratos
datatrain["F_ESTRATOVIVIENDA"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribución por estrato de vivienda")
plt.xlabel("Estrato")
plt.ylabel("Estudiantes")
plt.show()
#Top 15 departamentos con mayor cantidad de estudiantes
plt.figure(figsize=(10,6))
datatrain['E_PRGM_DEPARTAMENTO'].value_counts().head(15).sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 15 departamentos con más estudiantes')
plt.xlabel('Cantidad de estudiantes')
plt.ylabel('Departamento')
plt.tight_layout()
plt.show()


```
